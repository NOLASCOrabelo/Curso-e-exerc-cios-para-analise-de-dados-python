```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/0203_grafico_de_barras_simples.csv")
```

```python
df.head()
```

```python
df.sort_values(by= 'Award', ascending= True, inplace= True)
```

```python
cores_grafico=['#FF6347','#F08080', '#FF8C00', '#1E90FF', '#4682B4', '#191970', '#7CFC00', '#2E8B57', '#32CD32', '#BA55D3', '#9400D3','#FF1493', '#F8F8FF'  ]
fix,ax = plt.subplots(figsize=(15,8))
ax.bar (df['Film'], df['Award'], color= cores_grafico)
plt.xlabel('Filmes', fontsize=15)
plt.xticks(rotation= 90)
plt.title('Indicação ao Oscar de 2022', fontsize=20)
plt.ylabel('Votos', fontsize=15)
plt.show()
```
