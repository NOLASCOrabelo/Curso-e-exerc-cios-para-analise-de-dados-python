```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2,venn3, venn3_circles
```

```python
apenas_instagram = 500
apenas_tiktok = 800
apenas_facebook = 350
instagram_e_tiktok = 400
instagram_e_facebook = 150
tiktok_e_facebook = 200
todos_tres = 120
```

```python
subsets = (
    apenas_instagram, #A
    apenas_tiktok, #B
    apenas_facebook, #C
    instagram_e_tiktok, #A e B
    instagram_e_facebook, #A e C
    tiktok_e_facebook, #B e C
    todos_tres, # A, B e C
)
```

```python
venn2(
subsets=(apenas_instagram, apenas_tiktok, instagram_e_tiktok),
set_labels=('Instagram', 'Tiktok')
)
plt.show()
```

```python
venn3(subsets=(apenas_instagram, apenas_tiktok, apenas_facebook, instagram_e_tiktok, instagram_e_facebook, tiktok_e_facebook, todos_tres), set_labels=('Instagram', 'Tiktok', 'Facebook'), set_colors=('Gray', 'Black', 'Blue'))
venn3_circles(subsets=(apenas_instagram, apenas_tiktok, apenas_facebook, instagram_e_tiktok, instagram_e_facebook, tiktok_e_facebook, todos_tres), linestyle='dashdot', color='Yellow')
plt.show()
```
